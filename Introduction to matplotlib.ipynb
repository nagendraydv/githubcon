```python
heart_desease=pd.read_csv("heart_desease.csv")
```

```python
heart_desease.head()
```

```python
heart_desease.plot.bar(x="chol",y="age")
```

```python
heart_desease.plot.scatter(x="chol",y="age",)
```

```python
over_50=heart_desease[heart_desease["age"]>50]
over_50.head(10)
```

```python
over_50.plot(kind="scatter",x="age",y="chol",c="target",figsize=(15,6));
```

```python
from matplotlib.pyplot import legend
fig,ax=plt.subplots(figsize=(10,10))
scatter =ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
ax.set(title="heart_desease_calculation",
      xlabel="age",
      ylabel="cholestral")
ax.legend( title="target");
ax.axhline(y=over_50["chol"].mean(),linestyle="--",c="red")
```

```python
over_50.head(5)
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10))
# plot the data
scatter =ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
# customized the data
ax0.set(title="heart_desease_calculation",
      xlabel="age",
      ylabel="cholestral")
# add the legend
ax0.legend( title="target");
#drawing horizontal line 
ax0.axhline(y=over_50["chol"].mean(),linestyle="--",c="red")

scatter =ax1.scatter(x=over_50["chol"],
                    y=over_50["thalach"],
                    c=over_50["target"])
#customized the data
ax1.set(title="heart_desease_heart_rate",
       xlabel="cholestral",
       ylabel="heart_beat")
# add the legend

ax1.legend(title="target")

#drawing horizontal line

ax1.axhline(y=over_50["thalach"].mean(),linestyle="--",c="blue")

fig.suptitle("heart_desease",fontsize=16,fontweight="bold")
```
